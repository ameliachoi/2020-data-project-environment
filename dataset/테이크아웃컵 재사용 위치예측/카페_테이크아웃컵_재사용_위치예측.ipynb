```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('소상공인시장진흥공단_상가(상권)정보_서울_20200630.csv', sep='|')
df.head()
```

```python
df['상권업종대분류명'].value_counts()
```

```python
# 대분류 음식만 분류하기
df = df[df['상권업종대분류명'] == '음식']
```

```python
# 중구 데이터만 분류하기 
junggu = df[df['도로명주소'].str.contains('중구')]
junggu.head(3)
```

```python
junggu.info()
```

```python
junggu['상권업종소분류명'].value_counts()
```

```python
# 카페 매장 순위로 정렬
pd.set_option('display.max_rows', 10)
junggu[junggu['상권업종소분류명'] == '커피전문점/카페/다방']['상호명'].value_counts().to_frame()
```

```python
# 갯수 상위 10위 카페
# '스타벅스', '이디야', '투썸플레이스', '공차', '커피빈', '엔제리너스', '할리스', '빽다방', '폴바셋', '탐앤탐스'
```

```python
imsi = junggu[junggu['상호명'].str.contains('스타벅스')]
list = ['이디야', '투썸플레이스', '공차', '커피빈', '엔제리너스', '할리스', '빽다방', '폴바셋', '탐앤탐스']

for n in range(9):
    imsi2 = junggu[junggu['상호명'].str.contains(list[n])]
    imsi = pd.concat([imsi, imsi2])

imsi.info()
```

```python
imsi['상호명'].value_counts()
```

```python
j_cafe = imsi
```

```python
# 우리가 필요한 칼럼은 상호명, 지점명, 도로명주소, 경도, 위도
j_cafe = j_cafe[['상호명', '지점명', '도로명주소', '경도', '위도']]
```

```python
j_cafe.to_csv('서울특별시_중구_카페_정보.csv', index=False, encoding='utf-8')
```
