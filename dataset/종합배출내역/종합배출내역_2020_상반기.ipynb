```python
import pandas as pd
import numpy as np
```

```python
trash_1 = pd.read_csv('종합배출내역(일별)목록_2020년_01월.csv', encoding='cp949')
trash_3 = pd.read_csv('종합배출내역(일별)목록_2020년_03월.csv', encoding='cp949')
trash_6 = pd.read_csv('종합배출내역(일별)목록_2020년_06월.csv', encoding='cp949')
trash_7 = pd.read_csv('종합배출내역(일별)목록_2020년_07월.csv', encoding='cp949')
trash_8 = pd.read_csv('종합배출내역(일별)목록_2020년_08월.csv', encoding='cp949')
```

```python
trash = pd.concat([trash_1, trash_3, trash_6, trash_7, trash_8], axis=0)
trash
```

```python
trash.to_csv('종합배출내역_2020_상반기.csv', encoding='utf-8', index=False)
```
