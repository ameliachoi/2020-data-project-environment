```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('전국_재활용센터_정보.csv')
data.head(3)
```

```python
# 선별이라는 단어가 들어가면 쓰레기 선별장으로 분류

data['재활용센터명'] = data['재활용센터명'].astype(str)
seon = data[data['재활용센터명'].str.contains('선별')]
recycle = data[data['재활용센터명'].str.contains('선별') == False]
```

```python
seon.shape, recycle.shape, data.shape
```

```python
province_seon = seon['제공기관명'].value_counts().to_frame()
province_seon.reset_index(drop=False, inplace=True)
province_seon
```

```python
province_recycle = recycle['제공기관명'].value_counts().to_frame()
province_recycle.reset_index(drop=False, inplace=True)
province_recycle
```

```python
province_seon.to_csv('전국_재활용선별장_map.csv', encoding='utf-8', index=False)
province_recycle.to_csv('전국_재활용센터_map.csv', encoding='utf-8', index=False)
```
