```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('전국재활용센터표준데이터.csv', encoding='cp949')
data.head(3)
```

```python
data.columns
```

```python
# 필요한 데이터만 골라내기 
data = data[['제공기관명','재활용센터명', '재활용센터운영구분', '소재지도로명주소', '위도', '경도']]
data.head(3)
```

```python
data['제공기관명'].value_counts()
```

```python
data['제공기관명'] = data['제공기관명'].astype(str).str[:5]
```

```python
# folium choropleth를 위해 데이터 정제
data['제공기관명'][data['제공기관명'].str.contains('경기도')] = '경기도'
data['제공기관명'][data['제공기관명'].str.contains('강원도')] = '강원도'
data['제공기관명'][data['제공기관명'].str.contains('제주')] = '제주특별자치도'
```

```python
data['제공기관명'] = data['제공기관명'].str.strip()
```

```python
province = data['제공기관명'].value_counts().to_frame()
province.reset_index(drop=False, inplace=True)
province
```

```python
data.to_csv('전국_재활용센터_정보.csv', encoding='utf-8', index=False)
province.to_csv('전국_재활용센터_map.csv', encoding='utf-8', index=False)
```
