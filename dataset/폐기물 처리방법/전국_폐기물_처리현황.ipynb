```python
# 데이터 정리, merge -> final dataset
cat_2016 = pd.read_csv('disposal_cat_2016.csv')
map_2016 = pd.read_csv('disposal_map_2016.csv')
money_2016 = pd.read_csv('disposal_money_2016.csv')
```

```python
# 데이터 정리하기 - column 명 변경 등
cat_2016.columns = cat_2016.iloc[1, :]
imsi = cat_2016.iloc[:, 1:5]
imsi['시도'] = cat_2016['시도(1)']
```

```python
imsi.drop([0, 1], axis=0, inplace=True)
imsi
```

```python
imsi.reset_index(drop=True, inplace=True)
```

```python
cat_2016 = imsi # 전체 배출량, 매립, 소각, 재활용
```

```python
map_2016.head()
```

```python
imsi2 = map_2016.iloc[:, 5:7]
imsi2 = pd.concat([map_2016['시도'], imsi2], axis=1)
imsi2.rename(columns = {
    '2016.3' : '생활폐기물 관리구역(면적, km2)',
    '2016.4' : '생활폐기물 관리구역(인구, 명)'
}, inplace=True)
imsi2.drop([0, 1], axis=0, inplace=True)
imsi2
```

```python
imsi2.reset_index(drop=True, inplace=True)
imsi2
```

```python
map_2016 = imsi2
```

```python
money_2016.head()
```

```python
imsi3 = money_2016.iloc[:, [2, 4, 7]]
imsi3.rename(columns = {
    '시도별(1)' : '시도'
}, inplace=True)
imsi3.reset_index(drop=True, inplace=True)
imsi3
```

```python
money_2016 = imsi3
```

```python
# pd.merge를 이용하여 df join

merge = pd.merge(cat_2016, map_2016, how='outer', on='시도')
merge = pd.merge(merge, money_2016, how='outer', on='시도')
merge
```

```python
merge.rename(columns = {'계' : '처리 총량'}, inplace=True)
```

```python
merge.head(3)
```

```python
merge = merge[['시도', '생활폐기물 관리구역(면적, km2)', '생활폐기물 관리구역(인구, 명)',
              '시설 설치비', '수집·운반등 처리비', '매립', '소각', '재활용', '처리 총량']]

merge.head()
```

```python
merge.to_csv('전국_폐기물_처리현황_2016.csv', encoding='utf-8', index=False)
```
