```python
ja_mae_16 = ja_mae.iloc[:, :10]
ja_mae_17 = ja_mae.iloc[:, 10:19]
ja_mae_18 = ja_mae.iloc[:, 19:]
ja_mae_18.head(3)
```

```python
ja_mae_17 = pd.concat([
    ja_mae['시도별(1)'], ja_mae_17
], axis=1)

ja_mae_18 = pd.concat([
    ja_mae['시도별(1)'], ja_mae_18
], axis=1)

ja_mae_18.head(3)
```

```python
ja_mae_16.columns = ja_mae_16.iloc[0, :]
ja_mae_17.columns = ja_mae_17.iloc[0, :]
ja_mae_18.columns = ja_mae_18.iloc[0, :]

ja_mae_16.drop(0, axis=0, inplace=True)
ja_mae_17.drop(0, axis=0, inplace=True)
ja_mae_18.drop(0, axis=0, inplace=True)

ja_mae_18.head(3)
```

```python
ja_mae_16.to_csv('자기처리업체_매립_2016.csv', encoding='utf-8', index=False)
ja_mae_17.to_csv('자기처리업체_매립_2017.csv', encoding='utf-8', index=False)
ja_mae_18.to_csv('자기처리업체_매립_2018.csv', encoding='utf-8', index=False)
```
