```python
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings('ignore')
```

```python
cd './recycle'
```

```python
import os

path = "."
file_list = os.listdir(path)
file_list_csv = [file for file in file_list if file.endswith(".csv")]

print ("file_list_csv: {}".format(file_list_csv))
```

```python
data = pd.DataFrame(columns=['시점', '품목별', '수도권', '강원', '충북', '충남', '전북', '전남', '경북', '경남'])

for i in file_list_csv:
    df = pd.read_csv(i, encoding='cp949')
    data = pd.concat([data, df], axis=0)

data.shape
```

```python
data
```

```python
data.drop('Unnamed: 10', axis=1, inplace=True)
data
```

```python
imsi = data[['시점', '품목별']]
sudo = data['수도권']

data.drop(['시점', '품목별', '수도권'], axis=1, inplace=True)
data = pd.concat([imsi, sudo, data], axis=1)
data
```

```python
recycle = data
recycle.reset_index(drop=True, inplace=True)
```

```python
recycle.info()
```

```python
# 전부 object여서 숫자는 정수로 바꿔야함 
# 일단 결측치 처리를 0으로 한 다음에 진행해야함
```

```python
pd.isna(recycle).sum()
```

```python
for c in recycle.columns:
    recycle.loc[pd.isna(recycle[c]), c] = 0 # 전부 0으로 대체
    
pd.isna(recycle).sum()
```

```python
# 0을 제외하고도 - 이라는 문자가 있음... 이게 뭐지...?
recycle['수도권'].value_counts()
```

```python
for c in recycle.columns:
    recycle.loc[(recycle[c] == '-'), c] = 0 # 전부 0으로 대체
```

```python
recycle['수도권'].value_counts()
```

```python
recycle
```

```python
for c in ['수도권', '강원', '경남', '경북', '전남', '전북', '충북', '충남']:
    recycle[c] = recycle[c].astype(str)
    recycle[c] = recycle[c].str.replace(',', '')
    recycle[c] = recycle[c].astype('float64')
```

```python
recycle.dtypes # 변경 완료
```

```python
# 시점이 통일되어 있지 않아서 변환 필요
recycle['시점'].value_counts()
```

```python
# ['01-Jan', 'Oct-19', '01-Feb', '25-Nov', '01-Dec', '01-Mar'] 데이터 바꾸기
recycle['시점'] = recycle['시점'].astype(str)

recycle['시점'] = recycle['시점'].str.replace('01-Jan', '2020-01')
recycle['시점'] = recycle['시점'].str.replace('Oct-19', '2019-10')
recycle['시점'] = recycle['시점'].str.replace('01-Feb', '2020-02')
recycle['시점'] = recycle['시점'].str.replace('25-Nov', '2019-11')
recycle['시점'] = recycle['시점'].str.replace('01-Dec', '2019-12')
recycle['시점'] = recycle['시점'].str.replace('01-Mar', '2020-03')
```

```python
recycle['시점'] = recycle['시점'].str[:7]
recycle['시점'] = recycle['시점'].astype(object)
recycle['시점'].value_counts()
```

```python
recycle = recycle.sort_values(by='시점', axis=0, ascending=True)
recycle.reset_index(drop=True, inplace=True)
recycle.head(3)
```

```python
recycle.to_csv('재활용_자원_단가.csv', encoding='utf-8', index=False)
```
