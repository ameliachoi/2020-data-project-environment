```python
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns

import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('종합배출내역_2020_상반기.csv')
data.head()
```

```python
data.dismonth.value_counts()
```

```python
import calendar
```

```python
# 모든 axes에 동일한 scale 적용 : sharex=True, sharey=True
fig, axes = plt.subplots(nrows=6, figsize=(14,11), sharex=True, sharey=True)

for i, ax in enumerate(axes, 1):
    sns.kdeplot(data.query(f"dismonth=={i}")["disquantity"],
                shade=True, legend=False, color='tab:green', ax=ax)

    ax.set_xlim(1867911332 - (1000000000/3), 3580833676 + (1000000000/3))
    # ax.text(1867911332 - (1000000000/3), 15 , calendar.month_name[i], fontdict={"fontsize":"large"}) 
    
    ax.spines["left"].set_visible(False)
    ax.spines["right"].set_visible(False)
    ax.spines["top"].set_visible(False)
    if i != 6:
        ax.spines["bottom"].set_visible(False)
        ax.tick_params(axis="x", length=0)
    else:
        ax.tick_params(axis="x", labelsize="large")
        ax.set_xlabel("Quantity (g)", fontdict={"fontsize":"large"})
    
    ax.set_yticks([])    
    
    ax.grid(axis="x")
    ax.axhline(0, color="gray")
    ax.patch.set_alpha(0)
    
fig.subplots_adjust(hspace=-0.5)
```
